```python
import re, time, requests, pandas as pd
from pathlib import Path
from bs4 import BeautifulSoup
from tqdm.auto import tqdm

# ------------------------------------------------------------------
# CONFIG
# ------------------------------------------------------------------
PDF_DIR = Path("/Users/srinivasana/Documents/peds_agents/data/nda")          # destination folder
PDF_DIR.mkdir(parents=True, exist_ok=True)

INPUT_CSV = "/Users/srinivasana/Documents/peds_agents/data/interim/plc_candidates.csv"           # <-- change if you already have a DataFrame
OUTPUT_CSV = "/Users/srinivasana/Documents/peds_agents/data/interim/plc_with_pdfs.csv"

# ------------------------------------------------------------------
# Scraper helpers
# ------------------------------------------------------------------
HEADERS = {
    "User-Agent": "Mozilla/5.0 (pediatric-extrapolation-scraper)",
    "Accept": "text/html,application/pdf;q=0.9,*/*;q=0.8",
}
S = requests.Session()
S.headers.update(HEADERS)

def latest_label_url(app_id: str) -> str:
    """
    Given a 5‑/6‑digit application number (e.g. '206494'), return the URL of the
    most recent ...lbl.pdf. If none, return ''.
    """
    url = ("https://www.accessdata.fda.gov/scripts/cder/daf/index.cfm"
           f"?event=overview.viewDrugDetails&ApplNo={app_id}")
    try:
        html = S.get(url, timeout=8).text
    except requests.exceptions.RequestException:
        return ""
    soup = BeautifulSoup(html, "html.parser")
    tag = soup.find("a", href=re.compile(r"lbl\.pdf$", re.I))
    if not tag:
        return ""
    href = tag["href"]
    return "https://www.accessdata.fda.gov" + href if href.startswith("/") else href

def fetch_label_pdf(cell: str) -> str:
    """
    cell = contents of the 'FDA Application Number(s)' column.
    Returns local file path if download succeeds, else ''.
    """
    ids = sorted(set(re.findall(r"\d{5,7}", str(cell))))
    for app_id in ids:
        pdf_url = latest_label_url(app_id)
        if not pdf_url:
            continue
        try:
            r = S.get(pdf_url, timeout=15)
            if r.ok and r.headers.get("content-type", "").startswith("application/pdf"):
                out = PDF_DIR / f"{app_id}.pdf"
                out.write_bytes(r.content)
                return str(out)
        except requests.exceptions.RequestException:
            time.sleep(1)
    return ""

# ------------------------------------------------------------------
# Load PLC, download PDFs, save new CSV
# ------------------------------------------------------------------
df = pd.read_csv(INPUT_CSV)              # or use your in‑memory df

tqdm.pandas(desc="Downloading label PDFs")
df["pdf_path"] = df["FDA Application Number(s)"].progress_apply(fetch_label_pdf)

df.to_csv(OUTPUT_CSV, index=False)
print(f"✓  PDFs downloaded for {df['pdf_path'].astype(bool).sum()} "
      f"of {len(df)} rows → {OUTPUT_CSV}")
```

```python
import re, datetime, requests, pandas as pd
from pathlib import Path
from bs4 import BeautifulSoup
from tqdm.auto import tqdm

PDF_DIR = Path("/Users/srinivasana/Documents/peds_agents/data/nda")
PDF_DIR.mkdir(parents=True, exist_ok=True)

session = requests.Session()
session.headers.update({"User-Agent": "PEB label fetcher 0.4"})

YEARS = list(range(datetime.date.today().year, 2005, -1))
SUPPS = ["s000", "s005", "s010", "s011", "s020", "s024", "s030", "s040"]

def url_exists(url: str) -> bool:
    try:
        r = session.head(url, timeout=(4, 6), allow_redirects=True)
        return r.ok and r.headers.get("content-type","").startswith("application/pdf")
    except requests.exceptions.RequestException:
        return False

def fetch_latest_label(cell: str) -> str:
    # pull every 5‑7 digit number, clean _x000D_
    ids = re.findall(r"\d{5,7}", str(cell).replace("_x000D_", " "))
    for app_id in ids:
        for yr in YEARS:
            for supp in SUPPS:
                url = (f"https://www.accessdata.fda.gov/drugsatfda_docs/label/"
                       f"{yr}/{app_id}{supp}lbl.pdf")
                if url_exists(url):
                    try:
                        pdf = session.get(url, timeout=(4, 15))
                        if pdf.ok:
                            out = PDF_DIR / f"{app_id}.pdf"   # no supplement in filename
                            out.write_bytes(pdf.content)
                            return str(out)
                    except requests.exceptions.RequestException:
                        pass
            # if nothing in this year folder, continue to earlier year
    return ""   # nothing found for any ID in the cell


```

```python
# -------------------------------------------------------------------
# apply to DataFrame
# -------------------------------------------------------------------
df = pd.read_csv("/Users/srinivasana/Documents/peds_agents/data/interim/plc_candidates.csv")      # or use your in‑memory DataFrame

tqdm.pandas(desc="Download label PDFs")
df["pdf_path"] = df["FDA Application Number(s)"].progress_apply(fetch_latest_label)

df.to_csv("plc_with_pdfs.csv", index=False)
print("Saved", df["pdf_path"].astype(bool).sum(), "label PDFs → raw/fda/")
```
